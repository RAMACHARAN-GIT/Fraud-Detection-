```python
import streamlit as st
import pandas as pd
import joblib
```

```python
model = joblib.load('fraud_detection_pipeline.pkl')

st.title('Fraud Detection Prediction App')
st.markdown("Please enter the transaction amount")
st.divider()
transaction_type = st.selectbox('Transaction Type', ['PAYMENT','TRANSFER','CASH_OUT','DEPOSIT'])
amount = st.numbr_input("Amount", min_value=0.0, value=1000.0)
oldbalanceOrg = st.number_input('Old Balance (Sender)', min_values = 0.0, value=1000.0)
newbalanceOrig = st.number_input("New Balance (Sender)", min_value = 0.0, value = 1000.0)
oldbalanceDest = st.number_input("Old Balance (Receiver)",min_value=0.0, value= 1000.0)
newbalacneDest = st.number_input("New Balance (Receiver)",min_value=0.0, value= 1000.0)

if st.button("Predict"):
  input_data = pd.DataFrame({{
      "type": transaction_type,
      "amount":amount
      "oldbalanceOrg":oldbalanceOrg
      "newbalanceOrig":newbalanceOrig
      "oldbalanceDest":oldbalanceDest
      "newbalacneDest":newbalacneDest
  }})
  prediction = model.predict(input_data)[0]
  st.subheader({"Prediction: '{int(prediction)}'"})

  if prediction == 1:
    st.error("This transaction is fraud")
  else:
    st.success("This transaction is not fraud")
```
